```python
# Import Libraries
import pandas as pd
import numpy as np

# Read Park Data
park_data = pd.read_csv("parks_data.csv")

# Parks Analysis, counting parks 
parks_count = park_data['neighborhood'].value_counts().reset_index()
parks_count.columns = ['Neighborhood', 'Parks']

# Remove any neighborhood that have invalid data
parks_count = parks_count.dropna()
parks_count['ParksRank'] = parks_count['Parks'].rank(ascending=False)

# Sort data by rank
parks_count = parks_count.sort_values('ParksRank')
parks_count.head(20)
```

```python
import matplotlib.pyplot as plt

# Limit to 100 entries
parks_count = parks_count.nlargest(100, 'Parks')

# Use Median Income to Plot out Data
plot_data = parks_count[['Neighborhood','Parks']].set_index('Neighborhood')
plot_data = plot_data.sort_values('Parks', ascending=True)

# Create the bar plot
fig, ax = plt.subplots(figsize=(30, 10))
plot_data[['Parks']].plot(kind='bar', ax=ax)
ax.set_title("Neighborhood Rankings for Parks")
ax.set_xlabel("Neighborhood")
ax.set_ylabel("Rank")

# Display the plot
plt.tight_layout()
plt.show()
```
